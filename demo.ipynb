## Deploy Stable Diffusion 1.5 from Model Gardern

https://console.cloud.google.com/vertex-ai/publishers/google/model-garden/10?cloudshell=false
Latent text-to-image diffusion model capable of generating photo-realistic images given a text input

````python
from PIL import Image
import base64
import requests
from io import BytesIO
````

## Access Deployed Stable Diffusion Model Endpoint from Cloud Functions

### Run Functions Locally

from terminal run

 ```functions-framework --target http_function --port 8081```

````python
url = "http://localhost:8081"
````

````python
data = {"prompt": "a car in Picasso style"}

resp = requests.post(url=url,json=data)
image = Image.open(BytesIO(base64.b64decode(resp.text)))
display(image)
````

### Call Deployed Cloud Functions

from terminal run 

```gcloud functions deploy sd-demo --gen2 --region=us-west1 --runtime=python310 --source=<path-to-root-dir> --entry-point=http_function --trigger-http```

````python
url = "https://sd-demo-tfxf3pzq5a-uw.a.run.app"
````

````python
data = {"prompt": "a squirrel in Picasso style"}
resp = requests.post(url=url,json=data)
image = Image.open(BytesIO(base64.b64decode(resp.text)))
display(image)
````
